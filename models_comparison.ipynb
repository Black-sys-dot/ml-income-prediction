```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("C:\\Users\\ayank\\Downloads\\archive (6)\\adult.csv")
df.head()
```

```python
df=df.drop(columns=["fnlwgt"])
df=pd.get_dummies(df,columns=["workclass","education","marital-status","occupation","relationship","race","gender","native-country"])
df.head()
```

```python
np.sum(df.isna())
```

```python
x=df.drop(columns=["income"])
y=df["income"]

y=y==">50K"
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)

scaler=StandardScaler()

x_train=scaler.fit_transform(x_train)
x_test=scaler.transform(x_test)
```

Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix

model=LogisticRegression()
model.fit(x_train,y_train)

y_pred=model.predict(x_test)
accuracy=model.score(x_test,y_test)
confusion_mat=confusion_matrix(y_test,y_pred)
print(f"Accuracy: {accuracy}")
print(f"Confusion matrix: {confusion_mat}")
```

SVM

```python
from sklearn.svm import SVC

model=SVC(C=0.9,gamma=0.3)
model.fit(x_train,y_train)

y_pred=model.predict(x_test)
accuracy=model.score(x_test,y_test)
confusion_mat=confusion_matrix(y_test,y_pred)
print(f"Accuracy: {accuracy}")
print(f"Confusion matrix: {confusion_mat}")
```

Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier

model=DecisionTreeClassifier(max_depth=3)
model.fit(x_train,y_train)

y_pred=model.predict(x_test)
accuracy=model.score(x_test,y_test)
confusion_mat=confusion_matrix(y_test,y_pred)
print(f"Accuracy: {accuracy}")
print(f"Confusion matrix: {confusion_mat}")
```

Random Forest

```python
from sklearn.ensemble import RandomForestClassifier

model=RandomForestClassifier(
    n_estimators=200,
    max_depth=None,
    max_features="sqrt",
    bootstrap=True,
    random_state=42
)
model.fit(x_train,y_train)

y_pred=model.predict(x_test)
accuracy=model.score(x_test,y_test)
confusion_mat=confusion_matrix(y_test,y_pred)
print(f"Accuracy: {accuracy}")
print(f"Confusion matrix: {confusion_mat}")
```

XGBoost

```python
from xgboost import XGBClassifier

model=XGBClassifier(
    n_estimators=500,
    learning_rate=0.06,
    max_depth=None,
    subsample=0.8,
    colsample_bytree=0.8,
    random_state=42,
    use_label_encoder=False,
    eval_metric="logloss"
)
model.fit(x_train,y_train)

y_pred=model.predict(x_test)
accuracy=model.score(x_test,y_test)
confusion_mat=confusion_matrix(y_test,y_pred)
print(f"Accuracy: {accuracy}")
print(f"Confusion matrix: {confusion_mat}")
```
